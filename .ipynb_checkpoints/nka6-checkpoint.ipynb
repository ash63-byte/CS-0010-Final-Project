```python
import pandas as pd
%matplotlib inline
```

```python
data = pd.read_csv("fire_incidents.csv", index_col="neighborhood")
data
```

```python
del data['incident_type']
del data['address']
del data['alarm_time']
del data['alarms']
data
```

```python
del data['call_no']
data
```

```python
data_filtered=data.dropna()
data_filtered
```

```python
del data_filtered['ward']
data_filtered
```

```python
data_filtered
```

```python
popular_fires=data_filtered.groupby("neighborhood").count()
```

```python
popular_fires
```

```python
del popular_fires["latitude"]
```

```python
popular_fires
```

```python
ordered_fires=popular_fires.sort_values(by=['type_description'])
```

```python
ordered_fires
```

```python
ordered_fires.plot()
```

```python
import matplotlib.pyplot as plt
```

```python
ordered_fires.plot(x='neighborhood',y='type_description', kind='scatter')
plt.show()
```
