```python
import pandas as pd
import numpy as np

%matplotlib inline
import matplotlib.pyplot as plt

data = pd.read_csv("ArrestDataSet.csv", index_col="INCIDENTNEIGHBORHOOD", parse_dates=True)
del data['PK']
del data['AGE']
del data['GENDER']
del data['RACE']
del data['ARRESTTIME']
del data['INCIDENTZONE']
del data['INCIDENTTRACT']
del data['COUNCIL_DISTRICT']
del data['PUBLIC_WORKS_DIVISION']
del data['CCR']

data.head(10)

```

```python
del data['ARRESTLOCATION']
del data['INCIDENTLOCATION']
del data['X']
del data['Y']
data_filtered = data.dropna()

data.groupby('INCIDENTNEIGHBORHOOD').count().nsmallest(15,'OFFENSES',keep='all')
```

```python
data.groupby('INCIDENTNEIGHBORHOOD').count().nsmallest(10,'OFFENSES',keep='all').plot.bar(y="OFFENSES", rot= 90, use_index = True)

plt.title("Neighborhood Offenses")
plt.xlabel("Number of Offenses")
plt.ylabel("Neighborhood")
plt.style.use('classic')
my_colors = [(10.0, 20.0, 0.75)]
plt.plot(kind='bar', stacked=True, color=my_colors)



```
